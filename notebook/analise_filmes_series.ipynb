```python
#Variaveis 

notas = [9.5, 8.7, 9.0, 8.5, 7.8, 9.8, 8.9, 7.5, 9.2, 8.0]

filmes = [
    "Vingadores",
    "Interestelar",
    "Matrix",
    "Titanic",
    "Avatar",
    "Clube da Luta",
    "O Poderoso Chefao",
    "Jurassic Park",
    "Batman: O Cavaleiro das Trevas",
    "Forrest Gump"
]

melhorfilme = ""
melhornota = 0

piorfilme = ""
piornota = 10

soma = 0

contador=0

contandinho=0








print("===== LISTA DE FILMES =====")

for i in range(len(filmes)):
    print(f"{i+1}. {filmes[i]} / Nota: {notas[i]}")

# Melhor filme

for i in range(len(filmes)):
    if notas[i] > melhornota:
        melhornota = notas[i]
        melhorfilme = filmes[i]

# Pior filme

for i in range(len(filmes)):
    if notas[i] < piornota:
        piornota = notas[i]
        piorfilme = filmes[i]

# Média

for nota in notas:
    soma = soma + nota

media = soma / len(notas)

# Relatório Final

print("\n===== RELATÓRIO FINAL =====")

print(f"Quantidade de filmes cadastrados: {len(filmes)}")

print(f"\n🎬 Melhor filme: {melhorfilme}")
print(f"⭐ Melhor nota: {melhornota}")

print(f"\n🎬 Pior filme: {piorfilme}")
print(f"⭐ Pior nota: {piornota}")



for i in range (len(notas)):
    if notas [i] > 9:
           contador +=1

print(f"\nQuantidade de Filmes com nota acima de 9: {contador}")



for i in range (len(notas)):
      if notas [i] < 7:
           contandinho +=1

print(f"\nQuantidade de Filmes com nota abaixo de 7: {contandinho}")


print(" \n🎬 Filmes acima da média: ")

for i in range(len(filmes)):
    if notas[i] > media:
        print(f"{filmes[i]} - Nota: {notas[i]}")




print("\n🎬 Filmes abaixo da média:")

for i in range(len(filmes)):
    if notas[i] < media:
        print(f"{filmes[i]} - Nota: {notas[i]}")

print(f"\n📊 Média geral das notas: {media:.2f}")













```
